```python
import py7zr
import os

target_dir = 'kkbox_data'
files = [f for f in os.listdir(target_dir) if f.endswith('.7z')]

for file in files:
    file_path = os.path.join(target_dir, file)
    with py7zr.SevenZipFile(file_path, mode='r') as z:
        z.extractall(path=target_dir)
    print(f"{file} 해제 완료!")
```

```python
import pandas as pd

# 1. 파일 불러오기 (경로 설정: ./data/...)
# 파일명은 다운로드하신 이름에 맞춰 수정하세요 (예: demographics.csv, raw_data.csv)
df_demo = pd.read_csv('./gb/sports_gb_F.csv')
df_raw = pd.read_csv('./gb/sports_gb_A.csv')
df_info = pd.read_csv('./gb/sports_gb_total.csv')

# 2. 데이터 병합 (UserID를 기준으로 기질과 행동을 합침)
# 인구통계(기질) 데이터와 베팅 로그(행동) 데이터를 결합합니다.
```

```python
# fixed 합계
fixed_agg = df_demo.groupby("UserID").agg({
    "StakeF": "sum",
    "WinF": "sum",
    "BetsF": "sum"
}).reset_index()

# live 합계
live_agg = df_raw.groupby("UserID").agg({
    "StakeL": "sum",
    "WinL": "sum",
    "BetsL": "sum"
}).reset_index()

# merge
agg_rebuilt = fixed_agg.merge(live_agg, on="UserID", how="outer").fillna(0)
agg_rebuilt["StakeA"] = agg_rebuilt["StakeF"] + agg_rebuilt["StakeL"]
```

```python
compare = df_info.merge(agg_rebuilt, on="UserID", suffixes=("_file3", "_rebuilt"))

(compare["StakeF_file3"] - compare["StakeF_rebuilt"]).abs().max()
```
